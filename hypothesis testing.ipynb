```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm_notebook
import os 
```

```python
Folder = 'Data/'
file_list = sorted(os.listdir(Folder))
file_list
```

```python
import glob
#Make a filepath query
q = 'Data/**/final*.csv.gz'
file_list = sorted(glob.glob(q,recursive=True))
file_list
```

```python
#use glod to get complete Filepaths
pd.read_csv(file_list[0])
```

```python
# ## use a list comprehension to load in all files into 1 dataframe

df = pd.concat([pd.read_csv(file, lineterminator='\n') for file in file_list])
df = df.reset_index(drop=True)

df
```

```python
#remove the data with the ids equal 0
df = df.loc[df['imdb_id']!='0']
df
```

```python
#columns to exclude
drop_cols = ['backdrop_path','backdrop_path','original_title','overview',
                 'poster_path','status','tagline','id','homepage',
                 'production_countries','video','production_companies','spoken_languages',
            'original_language']
df = df.drop(columns=drop_cols)
df
```

```python
#check for duplicates
df.duplicated().sum()
```

```python
#check for missing values
df.isna().sum()
```

```python
# check again that it fixed 
df = df.dropna(subset=['release_date'])
df.isna().sum()
```
